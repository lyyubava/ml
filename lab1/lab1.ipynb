```python
import csv
import matplotlib.pyplot as plt
import numpy as np
def read_data(filename):
    vec_x = []
    vec_y = []
    with open(filename) as f:
        # omitting the first line that describes table
        f.readline()
        data = csv.reader(f,  delimiter=',')
    
        for d in data:
            vec_x.append(float(d[1]))
            vec_y.append(float(d[2]))
    res = {"X": np.array(vec_x), "Y" : np.array(vec_y)}
    return res
```

```python
def main(x, y, iters, learning_rate=0.01):
    weight = 0
    bias = 0
    N = x.shape[0]
    loss = 1/N*np.sum((y-(np.dot(x, weight)+bias)))
    for _ in range(iters):
        predicted_y = np.dot(x, weight) + bias
        error = predicted_y - y
        dw = 1/N * np.dot(x.T, error)
        db = 1/N*np.sum(error)
        weight -= learning_rate*dw
        bias -= learning_rate*db
        loss = 1/N*np.sum((y-(np.dot(x, weight)+bias)))
        print(f"w = {weight} \nbias = {bias}\nloss = {loss}\n")
    return weight, bias

weight_train, bias_train = main(x, y,  500, learning_rate=0.915)
```

```python
filename_test_set = 'lab_1_test.csv'
x_test = read_data(filename_test_set)["X"]
y_test = read_data(filename_test_set)["Y"]
weight_test, bias_test = main(x_test, y_test,  1200, learning_rate= 0.95)
```

```python
def visualize():

    max_val_train = max(x)
    max_val_test = max(x_test)
    max_val = float(np.maximum(max_val_test, max_val_train))

    min_val_train = min(x)
    min_val_test = min(x_test)
    min_val = float(np.minimum(min_val_train, min_val_test))

  
    y_min = bias_train + weight_train*min_val
    y_max = bias_train + weight_train*max_val
    plt.plot(x, y, 'ro', x_test, y_test, 'go', [min_val, max_val],[y_min, y_max])
    plt.show()
```

```python
visualize()
```
